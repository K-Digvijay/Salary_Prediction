```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('D:\TinkerForm\ML Flask_first app\hiring.csv')
df
```

```python
df['experience'].fillna(0,inplace=True)

df['test_score'].fillna(df['test_score'].mean(),inplace=True)
df
```

```python
def convert_word_to_int(word):
    word_dict = {
        'one':1,
        'two':2,
        'three':3,
        'four':4,
        'five':5,
        'six':6,
        'seven':7,
        'eight':8,
        'nine':9,
        'ten':10,
        'eleven':11,
        'twelve':12,
        'zero':0,0:0
    }
    return word_dict.get(word)
```

```python
X = df.iloc[:,:3]
```

```python
X['experience'] = X['experience'].apply(lambda x:convert_word_to_int(x))
X
```

```python
y = df.iloc[:,-1]
```

```python
linear = LinearRegression()

linear.fit(X,y)

pickle.dump(linear,open('model.pkl','wb'))
```
